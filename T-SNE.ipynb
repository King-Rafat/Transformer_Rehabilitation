```python
from scipy.interpolate import Akima1DInterpolator
import numpy as np
import matplotlib.pyplot as plt
fig, axs = plt.subplots(3, 1, figsize=(15, 3), constrained_layout = True, dpi = 300)
x3, y3 = df['Accuracy T'].dropna(), df['CO_2 log'].dropna()
x4, y4 = df['Accuract Random'].dropna(), df['CO_2 log.1'].dropna()
x5,y5 = df['Accuracy Tuned'].dropna(), df['CO_2 log.2'].dropna()
x6,y6 = df['Accuracy Random Shuff'].dropna(), df['Co_2 log'].dropna()
x7, y7 = df['Accuracy tuned Shuff'].dropna(), df['CO_2 log.3'].dropna()
akima3 = Akima1DInterpolator(x3, y3)
akima4 = Akima1DInterpolator(x4, y4)
akima5 = Akima1DInterpolator(x5, y5)
akima6 = Akima1DInterpolator(x6, y6)
akima7 = Akima1DInterpolator(x7, y7)
x3_new = np.linspace(min(x3), max(x3), 1000)
x4_new = np.linspace(min(x4), max(x4), 1000)
x5_new = np.linspace(min(x5), max(x5), 1000)
x6_new = np.linspace(min(x6), max(x6), 1000)
x7_new = np.linspace(min(x7), max(x7), 1000)
plt.subplot(131)
plt.plot(x3_new, akima3(x3_new), 'black', label='Teacher\n(ResNet18)', linestyle = 'dotted')
plt.plot(x5_new, akima5(x5_new), 'red', label='Student 1\n(MobileNetV2)',linestyle='dashed')
plt.plot(x7_new, akima7(x7_new), 'green', label='Student 2\n(ShuffleNetV2)',linestyle='dashed')
plt.plot(x4_new, akima4(x4_new), 'red', label='Our Student 1\n(MobileNetV2)')
plt.plot(x6_new, akima6(x6_new), 'green', label='Our Student 2\n(ShuffleNetV2)')
plt.xlabel("Accuracy (%)", fontsize = 15)
plt.ylabel("$LCO_2$", fontsize = 15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.title('(a)', loc = 'center', fontsize = 15, fontname="Times New Roman", fontweight='bold')
plt.legend(bbox_to_anchor=(-0.15, 0.97), ncol=1, frameon=False, fontsize = 9.6)
ax1 = plt.subplot(132)
ax1.bar(['Teacher\n(ResNet18)', 'Student 1\n(MobileNetV2)', 'Student 2\n(ShuffleNetV2)', 'Our Student 1\n(MobileNetV2)', 'Our Student 2\n(ShuffleNetV2)'], [4.7,2.76,2.72,2.79,2.68], width=0.35, 
        color=['white', 'white','white','green','red'], hatch = ['.','///','///',None,None], edgecolor=['black','green', 'red', 'black','black'], lw=1., zorder = 0)
ax1.bar(['Teacher\n(ResNet18)', 'Student 1\n(MobileNetV2)', 'Student 2\n(ShuffleNetV2)', 'Our Student 1\n(MobileNetV2)', 'Our Student 2\n(ShuffleNetV2)'], [4.7,2.76,2.72,2.79,2.68], width=0.35, 
        color='none', edgecolor='black', zorder=1, lw=2.)
# plt.bar(['Teacher\n(ResNet18)', 'Student 1\n(MobileNetV2)', 'Student 2\n(ShuffleNetV2)', 'Our Student 1\n(MobileNetV2)', 'Our Student 2\n(ShuffleNetV2)'], [4.7,2.76,2.72,2.79,2.68], width=0.35, 
#         color=['white', 'green','red','green','red'], hatch = ['.',None,None,None,None], edgecolor = 'black')
plt.ylabel('Time (s)', fontsize=15)
plt.xticks(fontsize=9.6, rotation=45, ha="right")
plt.yticks(fontsize=15)
plt.title('(b)', loc = 'center', fontsize = 15, fontname="Times New Roman", fontweight='bold')
ax1 = plt.subplot(133)
ax1.bar(['Teacher\n(ResNet18)', 'Student 1\n(MobileNetV2)', 'Student 2\n(ShuffleNetV2)', 'Our Student 1\n(MobileNetV2)', 'Our Student 2\n(ShuffleNetV2)'], [0.087,0.054,0.044,0.055,0.044], width=0.35, 
        color=['white', 'white','white','green','red'], hatch = ['.','///','///',None,None], edgecolor=['black','green', 'red', 'black','black'], lw=1., zorder = 0)
ax1.bar(['Teacher\n(ResNet18)', 'Student 1\n(MobileNetV2)', 'Student 2\n(ShuffleNetV2)', 'Our Student 1\n(MobileNetV2)', 'Our Student 2\n(ShuffleNetV2)'], [0.087,0.054,0.044,0.055,0.044], width=0.35, 
        color='none', edgecolor='black', zorder=1, lw=2.)

# plt.bar(['Teacher\n(ResNet18)', 'Student 1\n(MobileNetV2)', 'Student 2\n(ShuffleNetV2)', 'Our Student 1\n(MobileNetV2)', 'Our Student 2\n(ShuffleNetV2)'], [0.087,0.054,0.044,0.055,0.044], width=0.35, 
#         color=['white', 'white','white','green','red'], hatch = ['.','///','///',None,None], edgecolor = ['black','green', 'black', 'black','black'])
plt.ylabel('$CO_2 (g)$', fontsize=15)
plt.xticks(fontsize=9.6,rotation=45, ha="right")
plt.yticks(fontsize=15)
plt.title('(c)', loc = 'center', fontsize = 15, fontname="Times New Roman", fontweight='bold')
plt.show
plt.savefig('Final.pdf')
```
